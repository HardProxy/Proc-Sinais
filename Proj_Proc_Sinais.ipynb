##A resposta ao impulso será dada por meio da FFT inversa:

A FFT inversa para 30 pontos

```python
ifft30 = np.fft.ifft(df30.y)
plt.plot(ifft30)
```

A FFT inversa para 70 pontos

```python
ifft70 = np.fft.ifft(df70.y)
plt.plot(ifft70)
```

A FFT inversa para 150 pontos

```python
ifft150 = np.fft.ifft(df150.y)
plt.plot(ifft150)
```

Em cara uma das aplicações foi possivel observar que aumentando o numero de pontos presentes na amostra a tranformada rápida de Fourier tem valores cada vez mais precisos.

#Obtendo o filtro via FFT

Podemos assim calcular a FFT para observar a definição do filtro gerada por cada uma das FFT's inversas:

Para 30 pontos

```python
fft30 = np.fft.fft(ifft30)
plt.plot(fft30)
```

Para 70 pontos 

```python
fft70 = np.fft.fft(ifft70)
plt.plot(fft70)
```

Para 150 pontos 

```python
fft150 = np.fft.fft(ifft150)
plt.plot(fft150)
```

Aqui podemos observar que à medida que se aumenta o número de amostras afeta diretamente a definição do filtro.

#Conclusões 

Vimos que é possível construir a resposta do impulso de um filtro genérico a partir da FFT inversa, resposta ao impulso. Aplicando a FFT direta observamos o comportamento de melhor definição do filtro com o maior número de amostras. Portando, é possível concluir que à medida que aumentamos a quantidade de amostras a acurária do filtro é obtida de forma mais eficiente.
